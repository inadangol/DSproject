## 6. Time Series Analysis:

Analyze the arrival_date_year, arrival_date_month, arrival_date_week_number,
and arrival_date_day_of_month columns to identify seasonality or trends in
bookings over time.

```python
col_date=['arrival_date_year', 'arrival_date_month', 'arrival_date_week_number','arrival_date_day_of_month']
hotel_data['arrival_date_month'] = pd.to_datetime(hotel_data['arrival_date_month'], format='%B').dt.month

# Aggregate data by each time dimension
yearly_bookings = hotel_data.groupby('arrival_date_year').size()
monthly_bookings = hotel_data.groupby('arrival_date_month').size()
weekly_bookings = hotel_data.groupby('arrival_date_week_number').size()
daily_bookings = hotel_data.groupby('arrival_date_day_of_month').size()

# Plotting
plt.figure(figsize=(14, 10))

# Yearly bookings
plt.subplot(2, 2, 1)
yearly_bookings.plot(kind='line', marker='o')
plt.title('Yearly Bookings')
plt.xlabel('Year')
plt.ylabel('Number of Bookings')

# Monthly bookings
plt.subplot(2, 2, 2)
monthly_bookings.plot(kind='line', marker='o')
plt.title('Monthly Bookings')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')

# Weekly bookings
plt.subplot(2, 2, 3)
weekly_bookings.plot(kind='line', marker='o')
plt.title('Weekly Bookings')
plt.xlabel('Week Number')
plt.ylabel('Number of Bookings')

# Daily bookings
plt.subplot(2, 2, 4)
daily_bookings.plot(kind='line', marker='o')
plt.title('Daily Bookings')
plt.xlabel('Day of Month')
plt.ylabel('Number of Bookings')

plt.tight_layout()
plt.show()
```
